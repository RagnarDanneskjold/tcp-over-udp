```python
#House Keeping
%pylab inline
import matplotlib.pyplot as plt
```

```python
def load(file_name):
    with open(file_name, "r") as f:
        res = [ [float(i) for i in x.strip().split()] for x in f.readlines()]
    return res
def plot(data):
    plt.plot([x[0] for x in data], [x[1] for x in data])
```
