```python
#House Keeping
%pylab inline
import matplotlib.pyplot as plt
```

```python
def load(file_name):
    with open(file_name, "r") as f:
        res = [ [float(i) for i in x.strip().split()] for x in f.readlines()]
    return res
def plot(data):
    plt.plot([x[0] for x in data], [x[1] for x in data])
```

```python
t0 = load("trace_results/trace-10000_bytes-0")
t1 = load("trace_results/trace-10000_bytes-1")
t2 = load("trace_results/trace-10000_bytes-2")
c0 = load("cwnd_results/cwnd-10000_bytes-0")
c1 = load("cwnd_results/cwnd-10000_bytes-1")
c2 = load("cwnd_results/cwnd-10000_bytes-2")
```
